```python
import seaborn as sns
import random

sns.set_context("talk")

scores = []

with open("vuamc_novelty.csv") as f:
    f.readline()
    for l in f:
        scores.append(float(l.split(",")[-1]))
```

```python
import matplotlib.pyplot as plt

sns.distplot(scores, color="blue")
plt.xlabel("novelty score")
plt.yticks([], [])
plt.xticks([-1, -0.5, 0, 0.5, 1.0], [-1, -0.5, 0, 0.5, 1.0])
sns.despine(top=True, right=True, left=True)
plt.show()
```
